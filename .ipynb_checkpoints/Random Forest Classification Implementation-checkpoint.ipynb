#### Import libraries

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split,GridSearchCV, RandomizedSearchCV

from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from sklearn.metrics import precision_score, recall_score, accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Gathering 

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

### Data Preprocessing 

```python
df.isna().sum()
```

```python
df.dtypes
```

#### Exploratory Data Analysis

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Define subplot grid size
n_cols = 4  # number of subplots per row
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # automatically calculate rows
```

```python

# Create figure and axes
plt.figure(figsize=(15, 5 * n_rows))

# Loop through each numeric column and plot
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.hist(df[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title(col, fontsize=12)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Feature Engineering 

```python
df.columns
```

```python
x = df.drop("Outcome", axis = 1)
```

```python
y = df["Outcome"]
```

#### Model Training 

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred = rf_model.predict(x_test)
```

```python
y_pred
```

#### Model Evaluation 

```python
# On test data
cnf_matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix")
print(cnf_matrix)
print()
print()

print("Classification Report")
print(classification_report(y_test,y_pred))
```

```python
# On train data 
y_pred_train = rf_model.predict(x_train)
```

```python
cnf = confusion_matrix(y_train,y_pred_train)
print("Confusion Matrix")
print(cnf_matrix)
print()
print()

print("Classification Report")
print(classification_report(y_train,y_pred_train))
```

```python
# The model get overfitting so we need to perform the hyperparamter tuning 
```

### Hyperparameter Tuning of Random Forest Model

```python
rf_model = RandomForestClassifier()
```

```python
param_grid = {"criterion":["gini", "entropy"],
              "n_estimators": np.arange(50,200),
              "max_depth":np.arange(4,15),
              "min_samples_split":np.arange(4,20),
              "min_samples_leaf":np.arange(3,10),
              "oob_score":[True],
              "random_state":[11]
              }

rscv_rf_model = RandomizedSearchCV(rf_model,param_grid,cv=5)
rscv_rf_model.fit(x_train,y_train)
rscv_rf_model.best_estimator_
```

```python
new_rf_model = rscv_rf_model.best_estimator_
```

```python
y_pred = new_rf_model.predict(x_test)
con_matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix")
print(con_matrix)
print()
print()
print("Classification Report")
print(classification_report(y_test,y_pred))
```

```python
y_pred_train = new_rf_model.predict(x_train)

conf_matrix = confusion_matrix(y_train,y_pred_train)
print("Confusion Matrix:")
print(conf_matrix)
print()
print("Classification Report: ")
print(classification_report(y_train,y_pred_train))
```
